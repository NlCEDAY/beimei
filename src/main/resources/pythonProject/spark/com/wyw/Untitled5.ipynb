```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType
```

```python
ss = SparkSession.builder.getOrCreate()
```

```python
df = ss.read.format("csv").option("delimiter"," ").load("file:///opt/LifeExpentancy.txt") \
    .withColumn("Country",col("_c0")) \
    .withColumn("LifeExp",col("_c2")) \
    .withColumn("Region",col("_c4")) \
    .select(col("Country"),col("LifeExp"),col("Region"))
#df.show()
df.describe("LifeExp").show()
```

```python
import pandas as pd
pandas_df = pd.read_csv("file:///opt/products.csv",header=None,usecols=[1,2,5])
pandas_df
```

```python
spark_df = ss.createDataFrame(pandas_df)
spark_df.show()
```
